# API 사용 방법
1. sklearn 으로부터 적절한 estimator 클래스 임포트해서 모델의 클래스 선택
2. 클래스틑 원하는값으로 인스턴스화해서 모델의 하이퍼파라미터 선택
3. 데이터를 특징 배열과 대상 벡터로 배치
4. 모델 인스턴스의 fit() 메서드를 호출해 모델을 데이터에 적합
5. 모델을 새 데이터에 대해서 적용
    - 지도 학습: predict() 활용해 알려지지 않은 데이터에 대한 레이블 예측
    - 비지도 학습: transform()이나 predict()메서드를 사용해 데이터의 속성을 변환하거나 추론

# API 사용 예제

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use(['seaborn-whitegrid'])
```

```python
x = 10 * np.random.rand(50)
y = 2 * x + np.random.rand(50)

plt.scatter(x, y);
```

```python
# 1. sklearn 으로부터 적절한 estimator 클래스 임포트해서 모델의 클래스 선택
from sklearn.linear_model import LinearRegression
```

```python
# 2. 클래스틑 원하는값으로 인스턴스화해서 모델의 하이퍼파라미터 선택
model = LinearRegression(
    fit_intercept=True
)
model
```

```python
# 3. 데이터를 특징 배열과 대상 백터로 배치
X = x[:, np.newaxis]
```

```python
# 4. 모델 인스턴스의 fit() 메서드를 호출해 모델을 데이터에 적합
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# 5. 모델을 새 데이터에 대해서 적용
xfit = np.linspace(-1, 11)
Xfit = xfit[:, np.newaxis]
yfit = model.predict(Xfit)
```

```python
plt.scatter(x, y)
plt.plot(xfit, yfit, "--r");
```

# 예제 데이터 제공
1. 일반적으로 dict 구조
2. data: 특징 데이터 세트
3. target: 분류는 레이블, 회귀는 숫자 결과
4. target_names: 개별 레이블의 이름
5. feature_names: 특징 이름
6. DESCR: 데이터 세트설명 및 특징

```python
from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
print(diabetes.keys())
```

```python
print(diabetes.data)
```

```python
print(diabetes.target)
```

```python
print(diabetes.DESCR)
```

```python
print(diabetes.feature_names)
```
